```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt 
```

```python
#import nn stuff

from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
#make a regressiong

X, y = make_regression(n_samples=200, n_features=1, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

print(X.shape)
```

```python
print(X)
plt.plot(X)
plt.show()
```

```python
regr = MLPRegressor(random_state=1, max_iter=10000).fit(X_train, y_train)
```

```python
g = regr.predict(X_test[:])
```

```python
regr.score(X_test, y_test)
```

```python
print(g)
plt.plot(g, label="prediction")
plt.plot(y_test, label="ground truth")
plt.legend()
plt.show()
```

```python
#make a sine wave
numOfSamples = 200

x = np.linspace(0, 50, numOfSamples)
y = np.sin(x)
print(y.shape)
g = np.array([np.multiply(y, 2)])
d =g.reshape(numOfSamples, 1)
print(d.shape)

#print(d)

plt.plot(y, label = "output space")
plt.plot(d, label = "input space")
plt.legend()
#plt.plot(inputSpace[:][0])
plt.show()
```

```python
#make prediction
#X_train, X_test, y_train, y_test = train_test_split(d, y, random_state=1)
XSINTRAIN, XSINTEST, YSINTRAIN, YSINTREST = train_test_split(d, y, random_state=1)

newRegr = MLPRegressor(random_state = 1, max_iter=500).fit(XSINTRAIN, YSINTRAIN)
```

```python
output = newRegr.predict(XSINTEST)
```

```python
newRegr.score(XSINTEST, YSINTREST)
```

```python
#print(output)
#plt.plot(output, label="prediction")
plt.plot(XSINTEST, label="ground truth")
plt.legend()
plt.show()
```
